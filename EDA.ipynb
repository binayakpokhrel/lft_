```python
import pandas as pd
import matplotlib
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Starting Things with importing the necessary modules


```python
all_data=pd.read_csv(".ipynb_checkpoints/train.csv",)
all_data.describe()

```

```python
print("The percentage of the 'NaN' Cabin data was %0.2f "%((all_data["Cabin"].isnull().sum()/len(all_data))*100.0))
```

#### ["Age"] is also a column with some 'NaN' values and ["Embarked"] also contains 2 of the 'NaN' in it,Other values do not seem to carry 'NaN


```python
all_data=all_data.drop(["Cabin"],axis=1)
all_data=all_data.dropna()
```


# Getting the data that we need


The Column ["Cabin"] consisted of the most 'NaN' values which limited the Dataset size if used 'dropna'.So, We got rid of the column itself at first before dropping all 'NaN' valued rows.


```python
display((all_data.Survived==0).sum())
```

```python
display((all_data.Survived==1).sum())
```

```python
all_data.Sex=all_data.Sex.replace('male',1)
all_data.Sex=all_data.Sex.replace('female',0)
```

# TheData we have denotes 'male' for Males and 'female' for Females in the column "Sex" , Converting them into 0 and 1 would make the most sense in manipulation!



# 'Survived' is the main target prediction!


- While some of the Data like "Name" and "PassengerID" dont seem to play any role to the outcome,

- Whereas Age,Sex,Pclass,Parch,Sibsp seem to influence the result in their own ways.

```python
features=list(all_data.columns.values)
for i in features:
    if all_data.shape[0] == len(pd.Series(all_data[i]).unique()) :
        features.remove(i)
sns.set(style="darkgrid",context="talk")
sns.pairplot(all_data[features], hue="Survived")
```

# In this Visualization We plotted The hue for targeted prediction('Survived') across all features so, that we get the direction we should look out for.

```python
all_data.Embarked=all_data.Embarked.replace("C",1)
all_data.Embarked=all_data.Embarked.replace("S",2)
all_data.Embarked=all_data.Embarked.replace("Q",3)
```

```python
all_data_corr=all_data[features].corr()
display(all_data_corr)
display(features)
sns.heatmap(all_data_corr)
```

```python
survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = all_data[all_data['Sex']==0]
men = all_data[all_data['Sex']==1]
ax = sns.distplot(women[women['Survived']==1].Age, label = survived, ax = axes[0], kde =False,color='Green')
ax = sns.distplot(women[women['Survived']==0].Age, label = not_survived, ax = axes[0], kde =False,color='Black')
ax.legend()
ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age, label = survived, ax = axes[1], kde = False)
ax = sns.distplot(men[men['Survived']==0].Age, label = not_survived, ax = axes[1], kde = False)
ax.legend()
form= ax.set_title('Male')
```

# It is an interesting graph comparing survived and not-survived population across 'Male' and 'Female' 
## 20-40 age-group males have died the most 
## while 20-40 age-group females have survived the most

```python
all_data['Marital']=[1 if "Mrs" in x else 0 if "Miss" in x else None for x in  all_data['Name']]
maritalData=all_data.dropna()
maritalData['Marital']=maritalData['Marital'].astype(int)
```

# Feature Extraction using Clues

## Here, I have tried to extract "Marital' status for females only from the "Title" column which possess the "Mrs" and "Miss" clues 

## This is to see if Marital Status plays any part in survival

```python
sns.heatmap(maritalData.corr())
```

# observations:-

## Married Women are more likely to be lower class than upper
## 

```python
sns.set(rc={'figure.figsize':(8.7,3.7)})

sns.barplot(x="Pclass",y="Survived",data=maritalData, hue='Marital')
```

# Upper class Females have near 95%  survival Chances regardless the marital status
# Lower class Females however have about 45% chances of Survival which is less than half the chances upper class have

```python

all_data["Page"]=all_data["Age"]*all_data["Pclass"]
```

## As, the age and Pclass seem to be highly correlated 
# Adding Features using Pclass and Age 


```python
sns.set(rc={'figure.figsize':(68.7,23.7)})
axes = sns.barplot(x='Page',y='Survived',data=all_data )
```

### With the increasing 'Page' there is trend of decreasing survival chances

### We can Say that the old age lower class(1) people had very limited chances of survival compared to higher(1) class young age counterparts.
